```python
import igraph as ig
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# given the graph g, I want to create the graph g4, where for each node i, I create a new node that connects only to that node.
g = ig.Graph.Lattice([20], nei=2)
g4 = g.copy()
for i in range(g.vcount()):
    g4.add_vertex()
    g4.add_edge(i, g4.vcount()-1)

# each new node has an attribute "happiness" assigned at random either 0 and 1
for i in range(g4.vcount()):
    if (i<g.vcount()):
        g4.vs[i]["shape"] = "circle"
    else:
        g4.vs[i]["shape"] = "square"
    g4.vs[i]["happiness"] = np.random.choice([0, 1])

# each old node has an attribute "happiness" obtained as the average of the "happiness" of the new nodes. weighted by the distance between it and them
    
for i in range(g.vcount()):
    g4.vs[i]["happiness"] = 0
    for j in range(g.vcount(), g4.vcount()):
        # I increase the happiness by the happiness of the new node, times the distance
        # d is the distance between node i and node j
        d = g4.shortest_paths(i, j)[0][0]
        g4.vs[i]["happiness"] += g4.vs[j]["happiness"]/d/5

B = g4.vs["happiness"]
cmap1 = LinearSegmentedColormap.from_list("vertex_cmap", ["blue", "red"])
color4 = [cmap1(v*0.99) for v in B]

ig.plot(g4, vertex_color=color4, vertex_size=20, edge_width=1, edge_curved=False, vertex_shape=g4.vs["shape"])
```
