```python
import igraph as ig
import pandas as pd
import numpy as np
from matplotlib.colors import LinearSegmentedColormap


cmap1 = LinearSegmentedColormap.from_list("vertex_cmap", ["blue", "red"])



BB = pd.read_csv('all_B.csv')
BB = BB.to_numpy()
B = BB[10,:]

g = ig.Graph.Read_Adjacency('ADJ.csv', mode = 'undirected')
for indx,v in enumerate(g.vs):
    v['value'] = B[indx]
B = g.vs['value']
color = [cmap1(v*0.99) for v in B]

ig.plot(g, vertex_color = color, vertex_size = 10, edge_width = 0.5, layout = g.layout('kk'), bbox = (0,0,300,300))
```
